```python
# Convert data types
globallandslides['ev_date'] = pd.to_datetime(globallandslides['ev_date'], errors='coerce')
globallandslides['ev_time'] = pd.to_datetime(globallandslides['ev_time'], format='%H:%M:%S', errors='coerce').dt.time
globallandslides['fatalities'] = pd.to_numeric(globallandslides['fatalities'], errors='coerce')
globallandslides['injuries'] = pd.to_numeric(globallandslides['injuries'], errors='coerce')
globallandslides['ev_imp_id'] = pd.to_numeric(globallandslides['ev_imp_id'], errors='coerce')
globallandslides['latitude'] = pd.to_numeric(globallandslides['latitude'], errors='coerce')
globallandslides['longitude'] = pd.to_numeric(globallandslides['longitude'], errors='coerce')
globallandslides['gaz_dist'] = pd.to_numeric(globallandslides['gaz_dist'], errors='coerce')
globallandslides['sub_date'] = pd.to_datetime(globallandslides['sub_date'], errors='coerce')
globallandslides['edit_date'] = pd.to_datetime(globallandslides['edit_date'], errors='coerce')
```

```python
duplicates = globallandslides.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Check for null values
null_counts = globallandslides.isnull().sum()
print("Missing values before handling:\n", null_counts)
```

```python
# Check for duplicates and remove them
globallandslides_cleaned = globallandslides.drop_duplicates()
print("Number of duplicate rows removed:", globallandslides.shape[0] - globallandslides_cleaned.shape[0])

# Drop columns with a high percentage of missing values if they are not essential
columns_to_drop = ['storm_name', 'photo_link', 'comments']
globallandslides_cleaned = globallandslides_cleaned.drop(columns=columns_to_drop)

# Fill missing values for specific columns
globallandslides_cleaned['src_link'].fillna('No link available', inplace=True)
globallandslides_cleaned['ev_time'].fillna('Unknown', inplace=True)
globallandslides_cleaned['ev_desc'].fillna('No description available', inplace=True)
globallandslides_cleaned['loc_desc'].fillna('Unknown location', inplace=True)
globallandslides_cleaned['ls_cat'].fillna('Unknown', inplace=True)
globallandslides_cleaned['ls_trig'].fillna('Unknown', inplace=True)
globallandslides_cleaned['ls_size'].fillna('Unknown', inplace=True)
globallandslides_cleaned['ls_setting'].fillna('Unknown', inplace=True)

# Drop rows with remaining missing values in crucial fields
crucial_columns = ['ev_title', 'loc_accu', 'ls_cat', 'ls_size', 'sub_date', 'ctry_name', 'ctry_code', 'div_name']
globallandslides_cleaned = globallandslides_cleaned.dropna(subset=crucial_columns)

# Verify cleaning and duplicates removal
missing_values_globallandslides_cleaned = globallandslides_cleaned.isnull().sum()
duplicate_rows_after = globallandslides_cleaned.duplicated().sum()

print("Missing values after cleaning:\n", missing_values_globallandslides_cleaned)
print("Number of duplicate rows after cleaning:", duplicate_rows_after)

# Save the cleaned dataset
globallandslides_cleaned.to_csv(r'C:\Users\srikr\OneDrive\Desktop\CleanedDataForDataManipulations\cleaned_globallandslides.csv', index=False)
print("Cleaned dataset saved as 'cleaned_globallandslides.csv'")
```

```python
# How has the frequency of landslides changed over the years?



import pandas as pd
import matplotlib.pyplot as plt

# Loading the cleaned dataset
globallandslides_cleaned = pd.read_csv(r'C:\Users\srikr\OneDrive\Desktop\CleanedDataForDataManipulations\cleaned_globallandslides.csv')

# Convert 'ev_date' to datetime
globallandslides_cleaned['ev_date'] = pd.to_datetime(globallandslides_cleaned['ev_date'], errors='coerce')

# Temporal Analysis: Frequency of landslides over the years
globallandslides_cleaned['year'] = globallandslides_cleaned['ev_date'].dt.year
yearly_landslides = globallandslides_cleaned['year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(yearly_landslides.index, yearly_landslides.values, marker='o')
plt.title('Frequency of Landslides Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Landslides')
plt.grid(True)
plt.show()
```

```python
# Which countries experience the most landslides?

# Geographical Analysis: Top 10 countries with most landslides
country_landslides = globallandslides_cleaned['ctry_name'].value_counts().head(10)

plt.figure(figsize=(10, 5))
country_landslides.plot(kind='bar')
plt.title('Top 10 Countries with Most Landslides')
plt.xlabel('Country')
plt.ylabel('Number of Landslides')
plt.show()
```

```python
# What are the most common triggers for landslides?

# Cause Analysis: Common triggers for landslides
cause_landslides = globallandslides_cleaned['ls_trig'].value_counts().head(10)

plt.figure(figsize=(10, 5))
cause_landslides.plot(kind='bar')
plt.title('Common Triggers for Landslides')
plt.xlabel('Trigger')
plt.ylabel('Number of Landslides')
plt.show()
```
